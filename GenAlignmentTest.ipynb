```python
import numpy as np
import matplotlib.pyplot as plt
import imageio as iio
```

```python
SIZE = 900
PPI = 300
DOT_SPACING_MIL = np.array([10, 20, 50, 100])
DOT_SPACING_IN = DOT_SPACING_MIL / 1000
```

```python
# Create a blank image
image = np.ones((SIZE, SIZE), dtype=np.uint8) * 255
```

```python
# Split the image into 4 views
for i in range(2):
    for j in range(2):
        x_view = int(i * SIZE // 2)
        y_view = int(i * SIZE // 2)
        view = image[i * SIZE // 2 : (i + 1) * SIZE // 2, j * SIZE // 2 : (j + 1) * SIZE // 2]
        # Set the dot spacing for this view
        # dot_spacing = round((DOT_SPACING_IN * PPI) * (i + 1))
        dot_spacing = round(DOT_SPACING_IN[i + j] * PPI)
        # Draw the dots
        for x in range(0, SIZE // 2):
            for y in range(0, SIZE // 2):
                if x % dot_spacing == 0 and y % dot_spacing == 0:
                    view[y, x] = 0
                # view[y, x] = 0
```
